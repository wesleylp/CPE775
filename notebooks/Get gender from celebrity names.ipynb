```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from collections import Counter
tokenizer = RegexpTokenizer(r'\w+')
```

```python
import wikipedia
```

```python
import requests, json

def getGenders(names):
    url = ""
    cnt = 0
    if not isinstance(names,list):
        names = [names,]
    
    for name in names:
        if url == "":
            url = "name[0]=" + name
        else:
            cnt += 1
            url = url + "&name[" + str(cnt) + "]=" + name

    req = requests.get("https://api.genderize.io?" + url)
    results = json.loads(req.text)

    retrn = []
    for result in results:
        if result["gender"] is not None:
            retrn.append((result["gender"], result["probability"], result["count"]))
        else:
            retrn.append((u'None',u'0.0',0.0))
    return retrn
```

```python
gender_words = ['he', 'his', 'she', 'her']
```

```python
gender_counter = {}
pages = {}
for name in names:
    if name not in gender_counter:
        gender_counter[name] = Counter()
        
    try:
        page = wikipedia.page(name)
    except (wikipedia.PageError, wikipedia.DisambiguationError) as e:
        gender_counter[name].update([getGenders(name)[0][0]])
        continue

    pages[name] = page
    bio = page.content
    bio_no_punc = tokenizer.tokenize(bio)
    word_count_dict = Counter(w.lower() for w in bio_no_punc if w.lower() in gender_words)
    gender_counter[name].update(word_count_dict)
```

```python
gender_counter
```

```python
guesses = {}

for name, counter in gender_counter.items():
    
    if 'None' in counter or len(counter) == 0:
        guesses[name] = 'undefined'
        continue
    
    if 'male' in counter:
        guesses[name] = 'male'
        continue
        
    if 'female' in counter:
        guesses['female'] = 'female'
        continue
        
    male = counter['he'] + counter['his']
    female = counter['she'] + counter['her']
    
    if male > female:
        guesses[name] = 'male'
    else:
        guesses[name] = 'female'
```

```python
for name, gender in guesses.items():
    if gender == 'undefined':
        print(name)
```
